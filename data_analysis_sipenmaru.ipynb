## Data analysis Sipenmaru

Data sudah di clean berdasarkan `asal tinggal pendaftar` dengan `sorting tanggal pendaftaran terakhir`

```python
import pandas as pd
```

```python
df = pd.read_csv("Pendaftar_asal.csv")
df
```

```python
# jangan tampilkan data yang Nan
df = df.dropna()
df
```

```python
# Ubah data alamat_kota menjadi Uppercase
df = df.copy()

df.loc[:, "alamat_kota"] = df["alamat_kota"].str.upper()

df
```

```python
# hitung jumlah berdasarkan kota yaitu yang paling sering mendaftar

hitung_kota =df["alamat_kota"].value_counts()
hitung_kota
```

```python
# ambil data 10 besar paling ramai pendaftar

hitung_kota.head(10)
```

```python
# statistik pakai matplotlib

import matplotlib.pyplot as plt
hitung_kota.head(10).plot.bar()
plt.show()
```

```python
# cari kota paling sedikit pendaftar

hitung_kota.tail(10)
```

```python
# statistik pakai matplotlib

import matplotlib.pyplot as plt
hitung_kota.tail(10).plot.bar()
plt.show()
```

```python
df
```

```python
# ambil data pendaftar hanya di tahun 2024 pada tgl_daftar

df["thn_daftar"] = pd.to_datetime(df["tgl_daftar"])
df["thn_daftar"] = df["thn_daftar"].dt.year
df
```

```python
thn2024 = df[df["thn_daftar"] == 2024]
thn2024
```

```python
# jumlah pendaftar 2024

hitung_2024 = thn2024["thn_daftar"].value_counts()
hitung_2024
```

```python
# hitung jumlah berdasarkan alamat_kota di tahun 2024
hitung_alamat_kota =thn2024["alamat_kota"].value_counts()
hitung_alamat_kota
```

```python
# buat matplotlib

import matplotlib.pyplot as plt
hitung_alamat_kota.plot.bar()
plt.show()
```
